```python
import sys
import pandas as pd
import torch
import sys
from tqdm import tqdm, trange

from bikebench.xml_handling import bcad_to_bikebench
from bikebench.resource_utils import datasets_path
from bikebench.data_loading import data_loading
```

```python
data_all = bcad_to_bikebench.build_bikebench_dataframe()
```

```python
display(data_all)
```

```python
from bikebench.transformation import one_hot_encoding
data_oh = one_hot_encoding.encode_to_continuous(data_all)
```

```python
from sklearn.model_selection import train_test_split
df_oh_train, df_oh_test, df_train, df_test = train_test_split(data_oh, data_all, test_size=0.2, random_state=42)
```

```python
df_oh_train.to_csv(datasets_path("Generative_Modeling_Datasets/bike_bench.csv"))
df_oh_test.to_csv(datasets_path("Generative_Modeling_Datasets/bike_bench_test.csv"))
df_train.to_csv(datasets_path("Generative_Modeling_Datasets/bike_bench_mixed_modality.csv"))
df_test.to_csv(datasets_path("Generative_Modeling_Datasets/bike_bench_mixed_modality_test.csv"))
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
data_train = data_loading.load_bike_bench_train()
data_test = data_loading.load_bike_bench_test()
data_all = pd.concat([data_train, data_test])

#render all
from bikebench.rendering import rendering
from bikebench import resource_utils

svg_path = resource_utils.datasets_path("Real_Extended_Data/svgs")
png_path = resource_utils.datasets_path("Real_Extended_Data/pngs")
xml_path = resource_utils.datasets_path("Real_Extended_Data/xmls")
renderer = rendering.RenderingEngine(number_rendering_servers = 1, server_init_timeout_seconds=120)

```

```python
import cairosvg
for i in trange(len(data_all)):
    try:
        row = data_all.iloc[i]
        res = renderer.render_clip(row)
        svg = res.image_bytes
        idx = row.name
        xml = res.xml_file
        with open(f"{svg_path}/{idx}.svg", "wb") as f:
            f.write(svg)
        png = cairosvg.svg2png(bytestring=svg)
        with open(f"{png_path}/{idx}.png", "wb") as f:
            f.write(png)
        with open(f"{xml_path}/{idx}.bcad", "w") as f:
            f.write(xml)
    except:
        print(f"Failed on {idx}")

```
